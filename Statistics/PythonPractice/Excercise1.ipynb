```python
import pandas
customers = pandas.read_csv("file_excercises/CustomerData_Merrimack.csv")
print(customers.columns)
```

```python
import numpy
import pandas
customers['CarBrand'].replace('-1', numpy.nan, inplace=True)
customers['CarOwnership'].replace('-1', numpy.nan, inplace=True)
series = numpy.sum(customers.isnull())
pandas.DataFrame(series).rename({0:'Count'}, axis="columns").query('Count > 0').sort_values('Count', ascending=False)
```
